```python
with open('gdrive/MyDrive/Colab_Notebooks/Images/GAN_with_Pinn/Params.txt', 'r') as f:
    lines = f.readlines()

epochs = []
current_epoch = ''

for line in lines:
    if line.startswith('Ep:'):
        if current_epoch:
            epochs.append(current_epoch)
        current_epoch = line
    else:
        current_epoch += line

epochs.append(current_epoch)
```

```python
import re
import numpy as np

# read in the text file
with open('gdrive/MyDrive/Colab_Notebooks/Images/GAN_with_Pinn/Params.txt', 'r') as f:
    file_contents = f.read()

# use regular expressions to extract the numerical values for each epoch
epoch_values = re.findall(r'\[([\d\s\.]+)\]', file_contents)

# convert the epoch values to numpy arrays
epoch_values = [np.fromstring(epoch, sep=' ') for epoch in epoch_values]

# print the epoch values
list = []
for epoch_num, values in enumerate(epoch_values):
    print(f'Epoch {epoch_num + 1}: {values}')
```
